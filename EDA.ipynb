```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("kleinanzeigen_immobilien.csv")
```

```python
df.head()
```

```python
sns.set_palette("pastel")
fig = plt.figure(figsize=(10, 5))
sns.scatterplot(x=range(len(df['price'])),y=df['price'])
plt.ylim(0, 2000000)
plt.xlabel('Anzahl der Immobilien')
plt.ylabel('Preis in Euro')
# Zeige den Plot an
plt.show()
```

```python
sns.set_palette("pastel")
fig = plt.figure(figsize=(10, 5))
sns.histplot(x=df['price'], bins=100, kde=True)
```

```python
df_numeric = df.select_dtypes(include=[float, int]).dropna(axis=1, how='all')


corr = df_numeric.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.show()
```
